```python
import numpy as np
import matplotlib.pyplot as plt
import random
from celluloid import Camera

import Classroom
```

```python
c = Classroom.classroom(50,50,0.1,1)

fig, ax = plt.subplots()

camera = Camera(fig)
for i in range(200):
    ax.imshow(c.arr)
    c.iterate()
    camera.snap()
animation = camera.animate()
animation.save('animate.gif', writer = 'imagemagick', fps=30)
```
